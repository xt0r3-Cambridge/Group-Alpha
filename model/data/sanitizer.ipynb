```python
import json
```

```python
keywords = []
headlines = []

with open('keywords', 'r') as file:
    keywords = [line.strip() for line in file.readlines()]

for filepath in ["dailymail", "guardian", "bbc", "karen-hao", "melissa_heikkila"]:
    labelled = []
    unlabelled = []
    with open(f"archive/{filepath}.json") as file:
        records = json.load(file)
        for record in records:
            record['main_heading'] = record['main_heading'].replace('  ', ' ')
            headline = record['main_heading']

            good = filepath in ["karen-hao", "melissa_heikkila"]
            for word in keywords:
                if word in headline:
                    good = True
                    break

            if good:
                headlines.append(record)
            elif filepath == 'guardian':
                print(headline)

    print(len(headlines))

with open('corrected_data.json', 'w') as file:
    json.dump(headlines, file)
```
